# (1) Apriori

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv("dataset_market_basket.csv")
```

```python
#menampilkan data

df
```

```python
#Menghitung total setiap item

df.sum()
```

```python
#Menampilkan kolom yang ada

df.columns
```

```python
#menentukan kombinasi item yang mungkin terjadi

from itertools import permutations #library
comb_all = list(permutations(df.columns, 2)) #menyimpan semua kemungkinan kombinasi barang
comb_all #menampilkan
```

```python
# Menentukan kombinasi dimana item yang ada saja yang dihitung

result = [] #simpan data ke bentuk list
for comb_item in comb_all: #mengulangi dataset
  if len(df[(df[comb_item[0]]==1) & (df[comb_item[1]]==1)]) >0: #Menentukan kombinasi ada atau tidak
    result.append(comb_item) #simpan hasil dalam bentuk tuple ke dalam list
result #tampilkan hasil
```

```python
#Menghitung jumlah kombinasi setiap itemset

finalresult ={} #simpan data ke dictionary
for item_s in result:
  banyak_data=len(df[
      (df[item_s[0]]==1) & (df[item_s[1]]==1)
      ])
  finalresult[str(item_s)] = banyak_data

finalresult
```

```python
# Menghitung Support

for item,ab in finalresult.items(): #mengulangi pembacaan dataset tiap kombinasi item
  
  #pengitungan support tiap kombinasi item
  support = ab/len(df)*100
  print(f"{item} support = {support:.3f}%")
```

```python
# Menghitung Confidence

for item,ab in finalresult.items(): 
  i = eval(item) 
  confidence = ab/df[i[0]].sum()*100 
  print(f"{i[0]}->{i[1]}  Confidence = {confidence:.1f}%") 
```

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('deteksi_uang_palsu.csv')
```

```python
# Menampilkan dataframe ke bentuk tabel

df1
```

```python
#Buat tabel jarak berdasarkan data baru yang ditambahkan

import numpy as np
databaru= pd.DataFrame({'diagonal': [171.50],
                         'height_left': [104.10],
                         'height_right': [104.20],
                         'margin_low': [5.60],
                         'margin_up': [3.57],
                         'length': [111.50]})
df1['distance'] = np.sqrt(
    (df1['diagonal'] - databaru['diagonal'].values[0])**2
    + (df1['height_left'] - databaru['height_left'].values[0])**2
    + (df1['height_right'] - databaru['height_right'].values[0])**2
    + (df1['margin_low'] - databaru['margin_low'].values[0])**2
    + (df1['margin_up'] - databaru['margin_up'].values[0])**2
    + (df1['length'] - databaru['length'].values[0])**2
)
print('Tabel Jarak : ') # Menampilkan tabel jarak
print(df1['distance'])
```

```python
#1. Tambahkan kolom jarak untuk menampilkan jarak dari dataset terhadap target prediksi

pd.set_option('display.max_columns', None) #menampilkan semua kolom sesuai baris
pd.set_option('display.width', 180) #mengatur lebar tabel

print(df1) #menampilkan
```

```python
#2. Urutkan data berdasarkan K jarak terdekat (Untuk K nya ambil akar 2 dari banyak data)

sorted = df1.sort_values('distance') #mengurutkan data kecil ke besar
k = int(np.sqrt(len(df1))) #Mengambil nilai K jarak terdekat
dataterdekat = sorted.head(k) #menentukan data terdekat
print(dataterdekat)
```

```python
#3. Tentukan apa hasil prediksinya berdasarkan label terbanyak

from sklearn.neighbors import KNeighborsClassifier
print("Hasil prediksinya adalah True")

#memisahkan data
x = df1.drop(columns=['is_genuine','distance'],axis=1)
y = df1['is_genuine']

trydata = databaru #inisiasi data baru menjadi data tes
knn = KNeighborsClassifier(n_neighbors=k)#membuat variabel knn yang berisikan algoritma KNN dengan nilai K = akar 2 dari banyak data
knn.fit(x, y)#melatih knn dengan data trainining
prediksi = knn.predict(trydata)#memprediksi hasil
print("Berdasarkan dari digunakannya library sklearn untuk memprediksi Algoritma KNN, hasilnya adalah "+ ' '.join(map(str, prediksi)))
```
